```python
import torch
from ssvep import EEGNet
from ssvep.data import preprocess_epoch
import numpy as np

model = EEGNet(n_chans=8, n_samples=500, n_classes=12)
model.eval()

epoch_raw = np.random.randn(8, 710).astype(np.float32)
epoch = preprocess_epoch(epoch_raw)

with torch.no_grad():
    x = torch.from_numpy(epoch[None, None, ...]).float()
    logits = model(x)

print("Logits shape:", logits.shape)
```
